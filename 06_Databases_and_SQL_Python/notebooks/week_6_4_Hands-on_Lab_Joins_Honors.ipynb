### Problem 1: List the case number, type of crime and community area for all crimes in community area number 18.

```python
%%sql
select ch.case_number, ch.primary_type, c.community_area_name
from chicago_crime_data ch inner join census_data c on ch.community_area_number = c.community_area_number
where ch.community_area_number = '18'
```

### Problem 2: List all crimes that took place at a school. Include case number, crime type and community name.

```python
%%sql
select ch.case_number, ch.primary_type, c.community_area_name
from chicago_crime_data ch left outer join census_data c on ch.community_area_number = c.community_area_number
where ch.location_description like '%SCHOOL%'
```
